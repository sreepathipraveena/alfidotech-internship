```python
""
AI Internship ‚Äì Task 1
Data Preprocessing Pipeline
Project: Breast Cancer Diagnosis Prediction
Author: sreepathi praveena
"""
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 1Ô∏è‚É£ Load dataset (built-in, no internet)
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

print("‚úî Dataset Loaded Successfully")

print("\nüîπ First 5 rows of dataset:")
display(df.head())

print("\nüîπ Dataset Shape:", df.shape)

# 2Ô∏è‚É£ Check Missing Values
print("\nüìå Missing Values per Column:")
print(df.isnull().sum())

# 3Ô∏è‚É£ Feature Scaling
scaler = StandardScaler()
X = df.drop("target", axis=1)
scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(scaled, columns=X.columns)
df_scaled["target"] = df["target"]

print("\n‚úî Feature Scaling Completed")

print("\nüîπ First 5 Rows After Preprocessing:")
display(df_scaled.head())

# 4Ô∏è‚É£ Train‚ÄìTest Split
X_train, X_test, y_train, y_test = train_test_split(
    df_scaled.drop("target", axis=1),
    df_scaled["target"],
    test_size=0.2,
    random_state=42
)

# 5Ô∏è‚É£ Save output files
df_scaled.to_csv("cleaned_dataset.csv", index=False)
X_train.to_csv("X_train.csv", index=False)
X_test.to_csv("X_test.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

print("\nüéâ TASK 1 Completed Successfully!")
print("üìÇ Files generated: cleaned_dataset.csv, X_train.csv, X_test.csv, y_train.csv, y_test.csv")
```

```python
# ===============================================================
# üöÄ AI Internship ‚Äî TASK-2 : MACHINE LEARNING MODEL
# Project: Breast Cancer Diagnosis Prediction using ML
# ===============================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Set style for plots
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# ===============================================================
print("üîç STEP 1 ‚Äî Loading Preprocessed Dataset")
# ===============================================================
df = pd.read_csv("cleaned_dataset.csv")

X = df.drop("target", axis=1)
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print("‚úî Dataset Loaded & Split Successfully")

# ===============================================================
print("\nüîç STEP 2 ‚Äî Training Machine Learning Models")
# ===============================================================

# Option A: Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_predictions = dt.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)

# Option B: Logistic Regression
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)
lr_predictions = lr.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_predictions)

# ===============================================================
print("\nüèÜ STEP 3 ‚Äî Model Performance Evaluation")
# ===============================================================
print(f"üå≥ Decision Tree Accuracy: {dt_accuracy:.4f}")
print(f"üìå Logistic Regression Accuracy: {lr_accuracy:.4f}")

# ===============================================================
# Confusion Matrix Visualization (Logistic Regression)
# ===============================================================
cm = confusion_matrix(y_test, lr_predictions)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="YlGnBu", cbar=True)
plt.title("Confusion Matrix - Logistic Regression", fontsize=16)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ===============================================================
# Bar Plot of Accuracy Comparison
# ===============================================================
models = ['Decision Tree', 'Logistic Regression']
accuracies = [dt_accuracy, lr_accuracy]

sns.barplot(x=models, y=accuracies, palette="Set2")
plt.ylim(0,1)
plt.title("Model Accuracy Comparison", fontsize=16)
plt.ylabel("Accuracy")
for i, v in enumerate(accuracies):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center', fontweight='bold')
plt.show()

# ===============================================================
# Optional: Decision Tree Visualization
# ===============================================================
plt.figure(figsize=(20,10))
plot_tree(dt, feature_names=X.columns, class_names=["Malignant","Benign"], filled=True, rounded=True)
plt.title("Decision Tree Visualization", fontsize=16)
plt.show()

print("\nüéâ TASK-2 COMPLETED ‚Äî Model Evaluation Finished!")
print("üí° Next Step: Model Optimization & Hyperparameter Tuning")
```
