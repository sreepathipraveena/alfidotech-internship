```python
""
AI Internship ‚Äì Task 1
Data Preprocessing Pipeline
Project: Breast Cancer Diagnosis Prediction
Author: sreepathi praveena
"""
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 1Ô∏è‚É£ Load dataset (built-in, no internet)
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

print("‚úî Dataset Loaded Successfully")

print("\nüîπ First 5 rows of dataset:")
display(df.head())

print("\nüîπ Dataset Shape:", df.shape)

# 2Ô∏è‚É£ Check Missing Values
print("\nüìå Missing Values per Column:")
print(df.isnull().sum())

# 3Ô∏è‚É£ Feature Scaling
scaler = StandardScaler()
X = df.drop("target", axis=1)
scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(scaled, columns=X.columns)
df_scaled["target"] = df["target"]

print("\n‚úî Feature Scaling Completed")

print("\nüîπ First 5 Rows After Preprocessing:")
display(df_scaled.head())

# 4Ô∏è‚É£ Train‚ÄìTest Split
X_train, X_test, y_train, y_test = train_test_split(
    df_scaled.drop("target", axis=1),
    df_scaled["target"],
    test_size=0.2,
    random_state=42
)

# 5Ô∏è‚É£ Save output files
df_scaled.to_csv("cleaned_dataset.csv", index=False)
X_train.to_csv("X_train.csv", index=False)
X_test.to_csv("X_test.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

print("\nüéâ TASK 1 Completed Successfully!")
print("üìÇ Files generated: cleaned_dataset.csv, X_train.csv, X_test.csv, y_train.csv, y_test.csv")
```
